```python
out_max = application['CNT_FAM_MEMBERS'].quantile(0.999)
out_min = application['CNT_FAM_MEMBERS'].quantile(0.001)
application = application[(application['CNT_FAM_MEMBERS']>out_min) & (application['CNT_CHILDREN']<out_max)] # Eliminando outliers em "CNT_FAM_MEMBERS"
```

#### 1.7 Definiação da regra de negócio

```python
records['STATUS'].replace({'C': 0, 'X' : 0}, inplace=True) # Substituindo valores "C" e "X" da coluna "STATUS" por 0, pois possuem a mesma equivalência (estão em dia)
records['STATUS'] = records['STATUS'].astype('int')
records['STATUS'] = records['STATUS'].apply(lambda x:1 if x >= 2 else 0) # Aplicando regra de corte para valores com "STATUS" maior que 2, ou seja, mais de 60 dias em atraso
records = records.groupby('ID').agg(max).reset_index() # Mantendo apenas um registro por ID, constando um status que apresenta o máximo de dias que este ID esteve inadimplente
```

#### 1.8 Juntando "application" com "records" em um novo dataframe

```python
df = application.join(records.set_index('ID'), on='ID', how='inner')
df.drop(['MONTHS_BALANCE'], axis=1, inplace=True)
```

# 2. Desenvolvendo o modelo

#### 2.1  Dividindo os dados entre variável dependente e classe categórica e teste e treino

```python
x = df.iloc[:,1:-1] # Colunas da variável dependente
y = df.iloc[:,-1] # Colunas da classe categórica
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

#### 2.2 Escalando dados para um resultado mais acurado

```python
mms = MinMaxScaler()
x_scaled = pd.DataFrame(mms.fit_transform(x_train), columns=x_train.columns)
x_test_scaled = pd.DataFrame(mms.transform(x_test), columns=x_test.columns)
```

#### 2.3 Lidando com o problema de superamostragem

```python
oversample = SMOTE()
x_balanced, y_balanced = oversample.fit_resample(x_scaled, y_train)
x_test_balanced, y_test_balanced = oversample.fit_resample(x_test_scaled, y_test)
```

#### 2.4 Definindo modelos a serem testados

```python
classifiers = {
    "LogisticRegression" : LogisticRegression(),
    "KNeighbors" : KNeighborsClassifier(),
    "SVC" : SVC(),
    "DecisionTree" : DecisionTreeClassifier(),
    "RandomForest" : RandomForestClassifier(),
    "XGBoost" : XGBClassifier()
}
```

#### 2.5 Executando os modelos e verificando acurácia

```python
train_scores = []
test_scores = []

for key, classifier in classifiers.items():
    classifier.fit(x_balanced, y_balanced)
    train_score = classifier.score(x_balanced, y_balanced)
    train_scores.append(train_score)
    test_score = classifier.score(x_test_balanced, y_test_balanced)
    test_scores.append(test_score)
    
classificadores = list(classifiers.keys())

colunas = ['Model', 'Train Score', 'Test Score']

# Criando uma figura e um eixo
fig, ax = plt.subplots()

table_data = list(zip(classificadores, train_scores, test_scores))
table = ax.table(cellText=table_data, colLabels=colunas, cellLoc='center', loc='upper left')

# Configurando a aparência da tabela
table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(2, 1.5)

for n in range(0,3):
    table.get_celld()[(0, n)].set_facecolor("lightblue")

# Removendo ticks e eixos da figura
ax.axis('off')
plt.show()
```

#### 2.6 Escolhendo o melhor modelo = XGBoost

```python
xgb = XGBClassifier()
model = xgb.fit(x_balanced, y_balanced)
prediction = xgb.predict(x_test_balanced)
```

#### 2.7 Resultado do modelo

```python
print(classification_report(y_test_balanced, prediction))
```
